# **Data Cleaning - Bereinigen von Tabellen**

Der allererste und einer der wichtigsten Schritte ist das Bereinigen von Daten. Fehlerhafte Daten führen zu Verzerrungen von Informationen und Unzuverlässigkeit in Modellen. Es gibt ein paar Klassiker, die man in Tabellen findet (z.B. Null-Values, Duplikate, Trailing/Leading Whitespace etc.), aber es kann auch echt weirde Unreinheiten geben, die auf kleine aber feine Fehler zurückzuführen sind. Natürlich ist die beste Taktik, gar nicht erst solche Fehler einzubauen. In den meisten Fällen sind Unreinheiten aber allein technisch bedingt unvermeidbar oder man hat gar keine Kontrolle über den Ursprung einer Tabelle.
